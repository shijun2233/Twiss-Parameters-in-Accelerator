Hill's equation:
$$
\frac{d^2 x }{ds^2} + Kx = 0
$$

$$
\frac{d^2 y }{ds^2} - Kx = 0
$$
其中，$K = \frac{q }{mv}g,g = \frac{\partial B_y}{\partial x} $,在$x$方向聚焦，则$K>0$，散焦反之，水平方向的聚焦对应垂直方向的散焦，若y方向聚焦，则$K<0$。\
飘移节的Hill方程的传输矩阵为：
$$
R = \begin{bmatrix}
  1& L\\
  0&1
\end{bmatrix}
$$
聚焦四级磁铁的Hill方程的传输矩阵为：
$$
R = \begin{bmatrix}
  \cos(\sqrt{K}L)& \frac{\sin(\sqrt{K}L)}{\sqrt{K}}\\
  -k\sin(\sqrt{K}L)&\cos(\sqrt{K}L)
\end{bmatrix}
$$,散焦四级磁铁的Hill方程的传输矩阵为：
$$
R = \begin{bmatrix}
  \cosh(\sqrt{K}L)& \frac{\sinh(\sqrt{K}L)}{\sqrt{K}}\\
  \sqrt{K}\sinh(\sqrt{K}L)&\cosh(\sqrt{K}L)
\end{bmatrix}
$$
Twiss参数与传输矩阵的关系为：
$$
\left[\begin{array}{c}
\beta_{2} \\
\alpha_{2} \\
\gamma_{2}
\end{array}\right]=\left[\begin{array}{ccc}
R_{11}^{2} & -2 R_{11} R_{12} & R_{12}^{2} \\
-R_{11} R_{21} & R_{11} R_{22}+R_{12} R_{21} & -R_{12} R_{22} \\
R_{21}^{2} & -2 R_{21} R_{22} & R_{22}^{2}
\end{array}\right]\left[\begin{array}{c}
\beta_{1} \\
\alpha_{1} \\
\gamma_{1}
\end{array}\right]
$$
发射度定义为：
$$
\gamma x^2 + 2\alpha x x'+\beta x'^2 = \varepsilon
$$
满足二维高斯分布的相椭圆为：
$$
\sigma_{22}x^2 - 2\sigma_{12}x x' + \sigma_{11}x'^2 = |\sigma|
$$
密度分布函数可写成如下形式：
$$
f(x,x') = \frac{1}{2\pi \sqrt{|\sigma |}} \exp [-\frac{1}{2|\sigma |}(\sigma_{22}x^2 - 2\sigma_{12}x x' + \sigma_{11}x'^2) ]
$$
其中$\sigma_{11} >0 ,\sigma_{22}>0, \sigma_{11}\sigma_{22}-\sigma_{12}^2 = |\sigma |$; \
束流特性矩阵为：
$$
\sigma=\varepsilon\left[\begin{array}{cc}
\beta & -\alpha \\
-\alpha & \gamma
\end{array}\right]
$$

从位置1到位置2的束流传输矩阵为M,则位置2处的束流特性矩阵为:
$$
\sigma_2 = M\sigma_1M^T
$$
